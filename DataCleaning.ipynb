```python
import pandas as pd
import os
from google.colab import drive
import datetime
drive.mount('/content/drive')

folder_path = '/content/drive/My Drive/Sales_snapshot_data'

dataframes = []

for filename in os.listdir(folder_path):
    if filename.endswith('.xlsx'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_excel(file_path)
        dataframes.append(df)
df = pd.concat(dataframes, ignore_index=True)
df
```

```python
data = df.drop_duplicates()
data
```

```python
new_data = data[['month', 'week', 'distribution_channel', 'sold_quantity', 'net_price']]
new_data
```

Modifying month column to be
make year and month parts and combine them in the format 'YYYY-MM'
drop month original columns

```python
new_data = new_data.copy()
new_data['month'] = new_data['month'].astype(str)


new_data['year'] = new_data['month'].str[:4]
new_data['month'] = new_data['month'].str[5:]
new_data
```


For the week column


```python
new_data['week'] = new_data['week'].astype(str)
new_data.loc[:, 'week'] = new_data['week'].str[4:6]
new_data
```

```python
def week_to_date(year, week):
    """Convert a year and week number to the starting date of that week."""
    d = datetime.date(year, 1, 1)
    if d.weekday() <= 3:
        d = d - datetime.timedelta(d.weekday())
    else:
        d = d + datetime.timedelta(7 - d.weekday())
    return d + datetime.timedelta(days=(week - 1) * 7)

# Ensure that year and week columns are integers
new_data['year'] = new_data['year'].astype(int)
new_data['week'] = new_data['week'].astype(int)

# Convert week numbers to dates
new_data['date'] = new_data.apply(lambda row: week_to_date(row['year'], row['week']), axis=1)

# Convert 'date' to datetime type
new_data['date'] = pd.to_datetime(new_data['date'])

# Extracting month and day from the date
new_data['month_day'] = new_data['date'].dt.strftime('%m-%d')

# Concatenate year, month, and day
new_data['full_date'] = new_data['year'].astype(str) + '-' + new_data['month_day']

# Display the DataFrame
print(new_data[['year', 'week', 'full_date']])
```

```python
column_order = ['date'] + [col for col in new_data.columns if col not in ['date']]
new_data = new_data[column_order]
new_data
```

```python
file_save = '/content/drive/My Drive/Sales_data.csv'
new_data.to_csv(file_save, index=False)
```
